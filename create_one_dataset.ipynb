```python
import pandas as pd
import os
import numpy as np
```

```python
PATH = os.getcwd() + "/csv/new_csv"
CSVS = [PATH + "/" + csv_file for csv_file in os.listdir(PATH)]

df1 = pd.read_csv(CSVS[0], sep=";")
df1.head()
```

```python
PATH = os.getcwd() + "/csv/old_csv"
CSVS = [PATH + "/" + csv_file for csv_file in os.listdir(PATH)]
```

```python
df1 = pd.read_csv(CSVS[0], sep=";")
del df1["country"]
del df1["model"]
df1['mileage'] = df1['mileage'].astype(str) + " km"
df1["currency"] = "EUR"
df1.head()
```

```python
df2 = pd.read_csv(CSVS[1], sep=";")
del df2["country"]
del df2["model"]
df2['mileage'] = df2['mileage'].astype(str) + " km"
df2["currency"] = "EUR"
df2.head()
```

```python
df3 = pd.read_csv(CSVS[2], sep=";")
del df3["country"]
del df3["model"]
df3['mileage'] = df3['mileage'].astype(str) + " km"
df3["currency"] = "PLN"
df3.head()
```

```python
df4 = pd.read_csv(CSVS[3], sep=";")
del df4["country"]
del df4["model"]
df4['mileage'] = df4['mileage'].astype(str) + " miles"
df4["currency"] = "GBP"
df4.head()
```

```python
df5 = pd.read_csv(CSVS[4], sep=";")
del df5["country"]
del df5["model"]
df5['mileage'] = df5['mileage'].astype(str) + " miles"
df5["currency"] = "USD"
df5.head()
```

```python
d = pd.concat([df1, df2], ignore_index=True)
d = pd.concat([d, df3], ignore_index=True)
d = pd.concat([d, df4], ignore_index=True)
d = pd.concat([d, df5], ignore_index=True)
d.head()
```

```python
d.shape
```

```python
df = pd.concat([df, d], ignore_index=True)
df.shape
```

```python
df.to_csv("csv/old_new_combined.csv", sep=",", index=False)
```
